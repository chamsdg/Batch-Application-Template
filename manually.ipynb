```python
#import the library
import pandas as pd
import pycaret
```

```python
#import dataset
data = pd.read_csv("Churn.csv", sep=',')
```

```python
data['Geography'].value_counts()
```

```python
target = "Exited"
```

```python
#create environnenemt pycaret
from pycaret.classification import setup
ex1 = setup(data, target, html = True, numeric_imputation = "zero", high_cardinality_features = ["Surname", "Geography"],
             (ordinal_features = {"Gender": ["Female", "Male"]})
```

```python
#Compares models
from pycaret.classification import compare_models
top5 = compare_models(n_select = 5, include = ["lr", "dt", "rf", "knn", "xgboost"])
```

```python
#tuned_model
from pycaret.classification import tune_model
tune_top5 = [tune_model(i) for i in top5]
```

```python
#blend models
from pycaret.classification import automl
best_model = automl(optimize = "Accuracy")
```

```python
best_model
```

```python
#finalize it
from pycaret.classification import finalize_model
final = finalize_model(best_model)
```

```python
print(final)
```

```python
from pycaret.classification import predict_model
predict_model(final, data)
```

```python
#save model
from pycaret.classification import save_model
save_model(final, "manuall code")
```

```python
#load model
from pycaret.classification import load_model
loading = load_model("manuall code")
```

```python
#predict on another dataset
test = predict_model(loading, data)
```
